### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Step 2. Create the DataFrame that should look like the one below.

```python
data = {
    "preTestScore": [4, 24, 31, 2, 3, 4, 24, 31, 2, 3],
    "postTestScore": [25, 94, 57, 62, 70, 25, 94, 57, 62, 70],
    "age": [20, 25, 30, 35, 40, 22, 28, 32, 36, 45],
    "sex": ["M", "F", "M", "F", "M", "F", "M", "F", "M", "F"]
}

df = pd.DataFrame(data)

df
```

### Step 3. Create a Scatterplot of preTestScore and postTestScore, with the size of each point determined by age
#### Hint: Don't forget to place the labels

```python
plt.scatter(
    df["preTestScore"],
    df["postTestScore"],
    s=df["age"] * 10
)

plt.xlabel("Pre Test Score")
plt.ylabel("Post Test Score")
plt.title("Pre Test Score vs Post Test Score")

plt.show()
```

### Step 4. Create a Scatterplot of preTestScore and postTestScore.
### This time the size should be 4.5 times the postTestScore and the color determined by sex

```python
colors = {"M":"blue", "F":"red"}

plt.scatter(
    df["preTestScore"],
    df["postTestScore"],
    s=df["postTestScore"] * 4.5,
    c=df["sex"].map(colors)
)

plt.xlabel("Pre Test Score")
plt.ylabel("Post Test Score")
plt.title("Score Comparison by Sex")

plt.show()
```
