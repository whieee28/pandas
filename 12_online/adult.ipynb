
# Adult dataset

- `education-num`: continuous.

- `race`: White, Asian-Pac-Islander, Amer-Indian-Eskimo, Other, Black.
- `sex`: Female, Male.
- `capital-gain`: continuous.
- `capital-loss`: continuous.
- `hours-per-week`: continuous.
- `native-country`: United-States, Cambodia, England, Puerto-Rico, Canada, Germany, Outlying-US(Guam-USVI-etc), India, Japan, Greece, South, China, Cuba, Iran, Honduras, Philippines, Italy, Poland, Jamaica, Vietnam, Mexico, Portugal, Ireland, France, Dominican-Republic, Laos, Ecuador, Taiwan, Haiti, Columbia, Hungary, Guatemala, Nicaragua, Scotland, Thailand, Yugoslavia, El-Salvador, Trinadad&Tobago, Peru, Hong, Holand-Netherlands.

### 1. (2đ) Import các thư viện cần thiết

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### 2. (2đ) Đọc dữ liệu từ file này  [address](https://github.com/thieu1995/csv-files/blob/main/data/pandas/adult.data) và gán tên biến là df. Đồng thời thiết lập tên các features dựa trên thông tin đã cho ở đầu bài.

```python
columns = [
    'age', 'workclass', 'fnlwgt', 'education', 'education-num',
    'marital-status', 'occupation', 'relationship', 'race', 'sex',
    'capital-gain', 'capital-loss', 'hours-per-week', 'native-country', 'salary'
]

df = pd.read_csv(
    'https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/adult.data',
    names=columns,
    skipinitialspace=True
)

display(df.head())
```

### 20. (6đ) Tỉ lệ người married hay single men (marital-status feature) chiếm nhiều hơn trong số những người kiếm được >50k? Những người có tình trạng hôn nhân (married) có marital-status feature bắt đầu với Married (Married-civ-spouse, Married-spouse-absent or Married-AF-spouse), còn lại được coi là single men.

```python
male_high_earners = df[(df['sex'] == 'Male') & (df['salary'] == '>50K')]

# Xác định người đã kết hôn
married_conditions = (
    male_high_earners['marital-status'].str.startswith('Married-civ-spouse') |
    male_high_earners['marital-status'].str.startswith('Married-spouse-absent') |
    male_high_earners['marital-status'].str.startswith('Married-AF-spouse')
)
married_men_count = male_high_earners[married_conditions].shape[0]

# Xác định người độc thân
single_men_count = male_high_earners[~married_conditions].shape[0]


print(f"Số lượng đàn ông kiếm >50K đã kết hôn: {married_men_count}")
print(f"Số lượng đàn ông kiếm >50K độc thân: {single_men_count}")

if married_men_count > single_men_count:
    print("Tỷ lệ đàn ông kiếm >50K đã kết hôn chiếm nhiều hơn.")
elif single_men_count > married_men_count:
    print("Tỷ lệ đàn ông kiếm >50K độc thân chiếm nhiều hơn.")
else:
    print("Tỷ lệ đàn ông kiếm >50K đã kết hôn và độc thân là ngang nhau.")
```

### 21. (6đ) Số giờ tối đa một người làm việc mỗi tuần (hours-per-week feature) là bao nhiêu? Có bao nhiêu người làm việc với số giờ như vậy và tỷ lệ những người kiếm được nhiều (>50K) trong số đó là bao nhiêu?

```python
max_hours_per_week = df['hours-per-week'].max()
print(f"Số giờ tối đa một người làm việc mỗi tuần là: {max_hours_per_week}")

people_working_max_hours = df[df['hours-per-week'] == max_hours_per_week]
count_people_max_hours = people_working_max_hours.shape[0]
print(f"Số người làm việc với số giờ tối đa ({max_hours_per_week} giờ) là: {count_people_max_hours}")

earners_over_50K_at_max_hours = people_working_max_hours[people_working_max_hours['salary'] == '>50K']
percentage_earners_over_50K_at_max_hours = (earners_over_50K_at_max_hours.shape[0] / count_people_max_hours) * 100

print(f"Tỷ lệ những người kiếm được nhiều (>50K) trong số đó là: {percentage_earners_over_50K_at_max_hours:.2f}%")
```

### 23. (10đ) Số người da trắng (white in race feature) kiếm được số tiền >50k ở nước Mỹ chiếm bao nhiêu phần trăm. Câu hỏi tương tự với người da đen (black in race feature).

```python
us_population = df[df['native-country'] == 'United-States']

# Người da trắng ở Mỹ
white_us_population = us_population[us_population['race'] == 'White']
white_us_earners_50K = white_us_population[white_us_population['salary'] == '>50K']

percentage_white_us_over_50K = (len(white_us_earners_50K) / len(white_us_population)) * 100
print(f"Phần trăm người da trắng ở Mỹ kiếm >50K là: {percentage_white_us_over_50K:.2f}%")

# Người da đen ở Mỹ
black_us_population = us_population[us_population['race'] == 'Black']
black_us_earners_50K = black_us_population[black_us_population['salary'] == '>50K']

percentage_black_us_over_50K = (len(black_us_earners_50K) / len(black_us_population)) * 100
print(f"Phần trăm người da đen ở Mỹ kiếm >50K là: {percentage_black_us_over_50K:.2f}%")
```
