# Apple Stock

### Introduction:

We are going to use Apple's stock price.


### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/appl_1980_2014.csv)

```python
url = "https://raw.githubusercontent.com/PacktPublishing/Pandas-Cookbook/master/data/appl_1980_2014.csv"
```

### Step 3. Assign it to a variable apple

```python
apple = pd.read_csv(url)

apple.head()
```

### Step 4.  Check out the type of the columns

```python
apple.dtypes
```

### Step 5. Transform the Date column as a datetime type

```python
apple["Date"] = pd.to_datetime(apple["Date"])

apple.dtypes
```

### Step 6.  Set the date as the index

```python
apple = apple.set_index("Date")

apple.head()
```

### Step 7.  Is there any duplicate dates?

```python
apple.index.duplicated().sum()
```

### Step 8.  Ops...it seems the index is from the most recent date. Make the first entry the oldest date.

```python
apple = apple.sort_index()

apple.head()
```

### Step 9. Get the last business day of each month

```python
apple_month = apple.resample("BM").last()

apple_month
```

### Step 10.  What is the difference in days between the first day and the oldest

```python
(apple.index[-1] - apple.index[0]).days
```

### Step 11.  How many months in the data we have?

```python
apple.index.to_period("M").nunique()
```

### Step 12. Plot the 'Adj Close' value. Set the size of the figure to 13.5 x 9 inches

```python
plt.figure(figsize=(13.5,9))

apple["Adj Close"].plot()

plt.title("Apple Stock Adjusted Close Price")
plt.xlabel("Date")
plt.ylabel("Adj Close")

plt.show()
```
